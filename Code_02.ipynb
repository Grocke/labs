# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
imagem_um = Image.open('/content/drive/My Drive/Colab Notebooks/bailey-zindel-NRQV-hBF10M-unsplash.jpg')
imagem_dois = Image.open('/content/drive/My Drive/Colab Notebooks/claudio-testa--SO3JtE3gZo-unsplash.jpg')
imagem_tres = Image.open('/content/drive/My Drive/Colab Notebooks/johannes-plenio-RwHv7LgeC7s-unsplash.jpg')

plt.figure(figsize=(20,4))

plt.subplot(131)
plt.title('Imagem 1')
plt.axis('off')
plt.imshow(imagem_um)

plt.subplot(132)
plt.title('Imagem 2')
plt.axis('off')
plt.imshow(imagem_dois)

plt.subplot(133)
plt.title('Imagem 3')
plt.axis('off')
plt.imshow(imagem_tres)

plt.show()

#subplot(4,3,10)

imagem_um_arr = np.array(imagem_um)
imagem_dois_arr = np.array(imagem_dois)
imagem_tres_arr = np.array(imagem_tres)

# Histograma 1

red = imagem_um_arr[:,:,0]
green = imagem_um_arr[:,:,1]
blue = imagem_um_arr[:,:,2]

plt.figure(figsize=(20,4))

plt.subplot(131)
plt.hist(red.ravel(), color='red')

plt.subplot(132)
plt.hist(green.ravel(), color='green')

plt.subplot(133)
plt.hist(blue.ravel(), color='blue')

# Histograma 2

red = imagem_dois_arr[:,:,0]
green = imagem_dois_arr[:,:,1]
blue = imagem_dois_arr[:,:,2]
plt.figure(figsize=(20,4))

plt.subplot(131)
plt.hist(red.ravel(), color='red')

plt.subplot(132)
plt.hist(green.ravel(), color='green')

plt.subplot(133)
plt.hist(blue.ravel(), color='blue')

# Histograma 3

red = imagem_tres_arr[:,:,0]
green = imagem_tres_arr[:,:,1]
blue = imagem_tres_arr[:,:,2]
plt.figure(figsize=(20,4))

plt.subplot(131)
plt.hist(red.ravel(), color='red')

plt.subplot(132)
plt.hist(green.ravel(), color='green')

plt.subplot(133)
plt.hist(blue.ravel(), color='blue')
```
